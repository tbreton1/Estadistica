**Justificación del uso de p = 0.5 si no hay datos preliminares:**

La fórmula para calcular el tamaño de muestra necesario para estimar una proporción poblacional es:

$ n = \frac{Z^2 \cdot p \cdot (1-p)}{ME^2} $

Donde:
- $n$ es el tamaño de muestra necesario.
- $Z$ es el valor crítico de la distribución Z para el nivel de confianza deseado.
- $p$ es la proporción poblacional estimada.
- $ME$ es el margen de error deseado.

Cuando no se tiene información previa sobre la proporción poblacional ($p$), se utiliza $p = 0.5$ (o 50%) en la fórmula. Esto se debe a que el producto $p \cdot (1-p)$ alcanza su valor máximo cuando $p = 0.5$.

Considera los siguientes ejemplos:
- Si $p = 0.1$, $p \cdot (1-p) = 0.1 \cdot 0.9 = 0.09$
- Si $p = 0.3$, $p \cdot (1-p) = 0.3 \cdot 0.7 = 0.21$
- Si $p = 0.5$, $p \cdot (1-p) = 0.5 \cdot 0.5 = 0.25$
- Si $p = 0.7$, $p \cdot (1-p) = 0.7 \cdot 0.3 = 0.21$
- Si $p = 0.9$, $p \cdot (1-p) = 0.9 \cdot 0.1 = 0.09$

Como puedes ver, el valor de $p \cdot (1-p)$ es máximo en 0.25 cuando $p$ es 0.5. Dado que este término está en el numerador de la fórmula del tamaño de muestra, usar $p = 0.5$ resulta en el **mayor tamaño de muestra posible** para un nivel de confianza y margen de error dados.

Al calcular el tamaño de muestra con $p = 0.5$, nos aseguramos de que el tamaño de muestra sea lo suficientemente grande como para cumplir con el margen de error deseado, independientemente de cuál sea la verdadera proporción poblacional. Es un enfoque conservador que garantiza que la estimación será lo suficientemente precisa.

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Problema: Prueba de Hipótesis para la Media (σ conocida)
# Contexto: Una empresa afirma que el peso promedio de sus paquetes de cereal es de 500 g (μ0).
# Se sospecha que la máquina está subllenando, basándose en una muestra de 40 paquetes con una media de 495 g (x̄).
# La desviación estándar poblacional (σ) es conocida y es de 15 g. Se utilizará un nivel de significancia (α) de 0.05.
# Objetivo: Determinar si hay suficiente evidencia estadística para concluir que la máquina está subllenando los paquetes de cereal.

print("### Problema: Prueba de Hipótesis para la Media (σ conocida)")
print("\n**Contexto:** Una empresa afirma que el peso promedio de sus paquetes de cereal es de 500 g (μ₀). Se sospecha que la máquina está subllenada, basándose en una muestra de 40 paquetes con una media de 495 g (x̄). La desviación estándar poblacional (σ) es conocida y es de 15 g. Se utilizará un nivel de significancia (α) de 0.05.")
print("\n**Objetivo:** Determinar si hay suficiente evidencia estadística para concluir que la máquina está subllenando los paquetes de cereal.")

# 1. Plantear las hipótesis nula y alternativa
# H0: μ = 500 (El peso promedio es de 500 g)
# H1: μ < 500 (El peso promedio es menor a 500 g - subllenado)
# Esta es una prueba de hipótesis unilateral (cola izquierda).
print("\n**1. Plantear las hipótesis nula y alternativa**")
print("H₀: μ = 500 (El peso promedio es de 500 g)")
print("H₁: μ < 500 (El peso promedio es menor a 500 g - subllenado)")
print("Esta es una prueba de hipótesis unilateral (cola izquierda).")

# 2. Establecer el nivel de significancia α
alpha = 0.05
print(f"\n**2. Establecer el nivel de significancia α**")
print(f"α = {alpha}")

# 3. Identificar el estadístico de prueba y su distribución
# Dado que conocemos la desviación estándar poblacional (σ) y n > 30, usamos el estadístico Z.
# Distribución bajo H0: Normal estándar Z.
# Fórmula del estadístico Z: Z = (x̄ - μ₀) / (σ / √n)
print("\n**3. Identificar el estadístico de prueba y su distribución**")
print("Dado que conocemos la desviación estándar poblacional (σ) y n > 30, usamos el estadístico Z.")
print("Distribución bajo H₀: Normal estándar Z.")
print("Fórmula del estadístico Z: Z = (x̄ - μ₀) / (σ / √n)")

# Datos proporcionados
n = 40        # Tamaño de la muestra
x_barra = 495 # Media muestral
sigma = 15    # Desviación estándar poblacional
mu_0 = 500    # Media hipotetizada (según H0)

# Calcular el error estándar
error_estandar = sigma / np.sqrt(n)

# 4. Determinar el valor crítico y la región de rechazo
# Para una prueba unilateral a la izquierda con α = 0.05, el valor crítico de Z deja el 5% del área en la cola izquierda.
# norm.ppf(alpha) nos da el valor Z tal que P(Z <= z_critico) = alpha
z_critico = norm.ppf(alpha)
print(f"\n**4. Determinar el valor crítico y la región de rechazo**")
print(f"Valor crítico de Z para α={alpha} (cola izquierda): {z_critico:.4f}")
print(f"Región de rechazo: Z_calculado ≤ {z_critico:.4f}")


# 5. Calcular el estadístico de prueba y el valor p
# Calcular el estadístico Z
z_estadistico = (x_barra - mu_0) / error_estandar
# Calcular el valor p para una prueba unilateral izquierda: P(Z <= z_estadistico)
valor_p = norm.cdf(z_estadistico)

print(f"\n**5. Calcular el estadístico de prueba y el valor p**")
print(f"Estadístico Z calculado: {z_estadistico:.4f}")
print(f"Valor p: {valor_p:.4f}")

# 6. Graficar la distribución, región de rechazo, valor crítico y estadístico calculado
print(f"\n**6. Graficar la distribución, región de rechazo, valor crítico y estadístico calculado**")
# Rango de valores para graficar la distribución normal estándar
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1) # PDF de la distribución normal estándar (media=0, desviación estándar=1)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Normal Estándar')

# Rellenar la región de rechazo (cola izquierda)
x_reject = np.linspace(-4, z_critico, 100)
y_reject = norm.pdf(x_reject, 0, 1)
plt.fill_between(x_reject, y_reject, color='red', alpha=0.5, label=f'Región de Rechazo (α={alpha})')

# Marcar el valor crítico de Z
plt.axvline(z_critico, color='red', linestyle='--', label=f'Valor Crítico Z = {z_critico:.4f}')

# Marcar el estadístico Z calculado
plt.axvline(z_estadistico, color='blue', linestyle='-', label=f'Estadístico Z Calculado = {z_estadistico:.4f}')

# Añadir etiquetas y título
plt.title('Distribución Normal Estándar con Región de Rechazo y Estadístico Z')
plt.xlabel('Valor de Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Tomar una decisión estadística y concluir en términos prácticos
print(f"\n**7. Tomar una decisión estadística y concluir en términos prácticos**")
print("Para tomar la decisión, comparamos el estadístico Z calculado con el valor crítico de Z, o el valor p con el nivel de significancia α.")

print(f"\nComparando estadístico Z vs. valor crítico:")
print(f"- Estadístico Z calculado: {z_estadistico:.4f}")
print(f"- Valor crítico Z: {z_critico:.4f}")
if z_estadistico <= z_critico:
    print(f"Como {z_estadistico:.4f} ≤ {z_critico:.4f}, el estadístico Z calculado cae en la región de rechazo. **Rechazamos la hipótesis nula (H₀)**.")
else:
    print(f"Como {z_estadistico:.4f} > {z_critico:.4f}, el estadístico Z calculado NO cae en la región de rechazo. **No rechazamos la hipótesis nula (H₀)**.")

print(f"\nComparando valor p vs. α:")
print(f"- Valor p: {valor_p:.4f}")
print(f"- α: {alpha}")
if valor_p <= alpha:
    print(f"Como {valor_p:.4f} ≤ {alpha}, el valor p es menor que α. **Rechazamos la hipótesis nula (H₀)**.")
else:
     print(f"Como {valor_p:.4f} > {alpha}, el valor p es mayor que α. **No rechazamos la hipótesis nula (H₀)**.")

print("\n**Conclusión en términos prácticos:**")
if z_estadistico <= z_critico:
    print(f"Con un nivel de significancia del {alpha*100}%, hay suficiente evidencia estadística para concluir que el peso promedio de los paquetes de cereal es significativamente menor a 500 g. Esto respalda la sospecha de que la máquina de llenado está descalibrada y subllenando los paquetes.")
else:
    print(f"Con un nivel de significancia del {alpha*100}%, no hay suficiente evidencia estadística para concluir que el peso promedio de los paquetes de cereal es significativamente diferente a 500 g.")

print("\n**Significado del valor p:**")
print(f"El valor p de {valor_p:.4f} significa que, si la máquina estuviera funcionando correctamente y el peso promedio real fuera de 500 g (es decir, si H₀ fuera verdadera), solo habría un {valor_p*100:.2f}% de probabilidad de observar una media muestral de {x_barra} g o menos simplemente por casualidad en una muestra de {n} paquetes.")
if valor_p <= alpha:
     print(f"Dado que esta probabilidad es baja (menor que nuestro umbral α de {alpha}), consideramos que el resultado observado es poco probable bajo H₀, lo que nos lleva a rechazarla.")
else:
    print(f"Dado que esta probabilidad es alta (mayor que nuestro umbral α de {alpha}), consideramos que el resultado observado es probable bajo H₀, lo que nos lleva a no rechazarla.")
```

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Problema: Prueba de Hipótesis para la Media (σ conocida) - Promedio en examen nacional
# Contexto: Un profesor afirma que el promedio de su clase en un examen nacional es diferente al promedio nacional de 75 puntos (μ0).
# Una muestra de 36 estudiantes de su clase tiene una media de 78 (x̄).
# La desviación estándar poblacional (σ) es conocida y es de 12.
# Se utilizará un nivel de significancia (α) de 0.05.
# Objetivo: Determinar si hay suficiente evidencia estadística para concluir que el promedio de la clase es diferente al promedio nacional.

print("### Problema: Prueba de Hipótesis para la Media (σ conocida) - Promedio en examen nacional")
print("\n**Contexto:** Un profesor afirma que el promedio de su clase en un examen nacional es diferente al promedio nacional de 75 puntos (μ₀). Una muestra de 36 estudiantes de su clase tiene una media de 78 (x̄). La desviación estándar poblacional (σ) es conocida y es de 12. Se utilizará un nivel de significancia (α) de 0.05.")
print("\n**Objetivo:** Determinar si hay suficiente evidencia estadística para concluir que el promedio de la clase es diferente al promedio nacional.")

# 1. Plantear las hipótesis nula y alternativa
# H0: μ = 75 (El promedio de la clase es igual al promedio nacional)
# H1: μ ≠ 75 (El promedio de la clase es diferente al promedio nacional)
# Esta es una prueba de hipótesis bilateral (de dos colas).
print("\n**1. Plantear las hipótesis nula y alternativa**")
print("H₀: μ = 75 (El promedio de la clase es igual al promedio nacional)")
print("H₁: μ ≠ 75 (El promedio de la clase es diferente al promedio nacional)")
print("Esta es una prueba de hipótesis bilateral (de dos colas).")

# 2. Establecer el nivel de significancia α
alpha = 0.05
print(f"\n**2. Establecer el nivel de significancia α**")
print(f"α = {alpha}")

# 3. Identificar el estadístico de prueba y su distribución
# Dado que conocemos la desviación estándar poblacional (σ) y n > 30, usamos el estadístico Z.
# Distribución bajo H0: Normal estándar Z.
# Fórmula del estadístico Z: Z = (x̄ - μ₀) / (σ / √n)
print("\n**3. Identificar el estadístico de prueba y su distribución**")
print("Dado que conocemos la desviación estándar poblacional (σ) y n > 30, usamos el estadístico Z.")
print("Distribución bajo H₀: Normal estándar Z.")
print("Fórmula del estadístico Z: Z = (x̄ - μ₀) / (σ / √n)")

# Datos proporcionados
n = 36       # Tamaño de la muestra
x_barra = 78 # Media muestral
sigma = 12   # Desviación estándar poblacional
mu_0 = 75    # Media hipotetizada (según H0)

# Calcular el error estándar
error_estandar = sigma / np.sqrt(n)

# 4. Determinar los valores críticos y la región de rechazo
# Para una prueba bilateral con α = 0.05, el α se divide en dos colas (α/2 = 0.025).
# Los valores críticos de Z son -Z(α/2) y +Z(α/2).
# norm.ppf(alpha/2) nos da el valor Z para la cola inferior.
# norm.ppf(1 - alpha/2) nos da el valor Z para la cola superior.
alpha_ci = alpha / 2
z_critico_inferior = norm.ppf(alpha_ci)
z_critico_superior = norm.ppf(1 - alpha_ci)

print(f"\n**4. Determinar los valores críticos y la región de rechazo**")
print(f"α/2 = {alpha_ci}")
print(f"Valores críticos de Z para α={alpha} (prueba bilateral): {z_critico_inferior:.4f} y {z_critico_superior:.4f}")
print(f"Región de rechazo: Z_calculado ≤ {z_critico_inferior:.4f} o Z_calculado ≥ {z_critico_superior:.4f}")


# 5. Calcular el estadístico de prueba y el valor p
# Calcular el estadístico Z
z_estadistico = (x_barra - mu_0) / error_estandar
# Calcular el valor p para una prueba bilateral: 2 * P(Z <= -|z_estadistico|) o 2 * P(Z >= |z_estadistico|)
# Usamos 2 * norm.cdf(-abs(z_estadistico)) para obtener el valor p correcto para prueba bilateral
valor_p = 2 * norm.cdf(-abs(z_estadistico))

print(f"\n**5. Calcular el estadístico de prueba y el valor p**")
print(f"Estadístico Z calculado: {z_estadistico:.4f}")
print(f"Valor p: {valor_p:.4f}")

# 6. Graficar la distribución, región de rechazo, valores críticos y estadístico calculado
print(f"\n**6. Graficar la distribución, región de rechazo, valores críticos y estadístico calculado**")
# Rango de valores para graficar la distribución normal estándar
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1) # PDF de la distribución normal estándar (media=0, desviación estándar=1)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Normal Estándar')

# Rellenar la región de rechazo (colas izquierda y derecha)
x_reject_inferior = np.linspace(-4, z_critico_inferior, 100)
y_reject_inferior = norm.pdf(x_reject_inferior, 0, 1)
plt.fill_between(x_reject_inferior, y_reject_inferior, color='red', alpha=0.5, label=f'Región de Rechazo Inferior (α/2={alpha_ci})')

x_reject_superior = np.linspace(z_critico_superior, 4, 100)
y_reject_superior = norm.pdf(x_reject_superior, 0, 1)
plt.fill_between(x_reject_superior, y_reject_superior, color='red', alpha=0.5, label=f'Región de Rechazo Superior (α/2={alpha_ci})')

# Marcar los valores críticos de Z
plt.axvline(z_critico_inferior, color='red', linestyle='--', label=f'Valor Crítico Z (Inf) = {z_critico_inferior:.4f}')
plt.axvline(z_critico_superior, color='red', linestyle='--', label=f'Valor Crítico Z (Sup) = {z_critico_superior:.4f}')

# Marcar el estadístico Z calculado
plt.axvline(z_estadistico, color='blue', linestyle='-', label=f'Estadístico Z Calculado = {z_estadistico:.4f}')

# Añadir etiquetas y título
plt.title('Distribución Normal Estándar con Región de Rechazo y Estadístico Z (Prueba Bilateral)')
plt.xlabel('Valor de Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Tomar una decisión estadística y concluir en términos prácticos
print(f"\n**7. Tomar una decisión estadística y concluir en términos prácticos**")
print("Para tomar la decisión, comparamos el estadístico Z calculado con los valores críticos de Z, o el valor p con el nivel de significancia α.")

print(f"\nComparando estadístico Z vs. valores críticos:")
print(f"- Estadístico Z calculado: {z_estadistico:.4f}")
print(f"- Valores críticos Z: {z_critico_inferior:.4f} y {z_critico_superior:.4f}")
if z_estadistico <= z_critico_inferior or z_estadistico >= z_critico_superior:
    print(f"Como {z_estadistico:.4f} ≥ {z_critico_superior:.4f}, el estadístico Z calculado cae en la región de rechazo. **Rechazamos la hipótesis nula (H₀)**.")
else:
    print(f"Como {z_estadistico:.4f} está entre {z_critico_inferior:.4f} y {z_critico_superior:.4f}, el estadístico Z calculado NO cae en la región de rechazo. **No rechazamos la hipótesis nula (H₀)**.")

print(f"\nComparando valor p vs. α:")
print(f"- Valor p: {valor_p:.4f}")
print(f"- α: {alpha}")
if valor_p <= alpha:
    print(f"Como {valor_p:.4f} ≤ {alpha}, el valor p es menor que α. **Rechazamos la hipótesis nula (H₀)**.")
else:
     print(f"Como {valor_p:.4f} > {alpha}, el valor p es mayor que α. **No rechazamos la hipótesis nula (H₀)**.")

print("\n**Conclusión en términos prácticos:**")
if valor_p <= alpha: # Usar valor p para la conclusión general para mayor claridad
    print(f"Con un nivel de significancia del {alpha*100}%, hay suficiente evidencia estadística para concluir que el promedio de la clase del profesor es significativamente diferente al promedio nacional de 75 puntos.")
else:
    print(f"Con un nivel de significancia del {alpha*100}%, no hay suficiente evidencia estadística para concluir que el promedio de la clase del profesor es significativamente diferente al promedio nacional de 75 puntos.")

print("\n**Significado del valor p:**")
print(f"El valor p de {valor_p:.4f} significa que, si el promedio real de la clase fuera de 75 puntos (es decir, si H₀ fuera verdadera), solo habría un {valor_p*100:.2f}% de probabilidad de observar una media muestral de {x_barra} puntos o algo más extremo (ya sea mayor o menor que 75) simplemente por casualidad en una muestra de {n} estudiantes.")
if valor_p <= alpha:
     print(f"Dado que esta probabilidad es baja (menor que nuestro umbral α de {alpha}), consideramos que el resultado observado es poco probable bajo H₀, lo que nos lleva a rechazarla.")
else:
    print(f"Dado que esta probabilidad es alta (mayor que nuestro umbral α de {alpha}), consideramos que el resultado observado es probable bajo H₀, lo que nos lleva a no rechazarla.")
```

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Problema: Prueba de Hipótesis para la Media (σ conocida) - Llenado de botellas
# Contexto: Una máquina llena botellas de 500 mL (μ0).
# Una auditoría revela que, en 60 botellas, la media es 498 mL (x̄).
# La desviación estándar poblacional (σ) es conocida y es de 5 mL.
# Se utilizará un nivel de significancia (α) de 0.01.
# Objetivo: Determinar si hay suficiente evidencia estadística para concluir que la media de llenado es diferente a 500 mL.

print("### Problema: Prueba de Hipótesis para la Media (σ conocida) - Llenado de botellas")
print("\n**Contexto:** Una máquina llena botellas de 500 mL (μ₀). Una auditoría revela que, en 60 botellas, la media es 498 mL (x̄). La desviación estándar poblacional (σ) es conocida y es de 5 mL. Se utilizará un nivel de significancia (α) de 0.01.")
print("\n**Objetivo:** Determinar si hay suficiente evidencia estadística para concluir que la media de llenado es diferente a 500 mL.")

# 1. Plantear las hipótesis nula y alternativa
# H0: μ = 500 (La media de llenado es de 500 mL)
# H1: μ ≠ 500 (La media de llenado es diferente a 500 mL)
# Esta es una prueba de hipótesis bilateral (de dos colas).
print("\n**1. Plantear las hipótesis nula y alternativa**")
print("H₀: μ = 500 (La media de llenado es de 500 mL)")
print("H₁: μ ≠ 500 (La media de llenado es diferente a 500 mL)")
print("Esta es una prueba de hipótesis bilateral (de dos colas).")

# 2. Establecer el nivel de significancia α
alpha = 0.01
print(f"\n**2. Establecer el nivel de significancia α**")
print(f"α = {alpha}")

# 3. Identificar el estadístico de prueba y su distribución
# Dado que conocemos la desviación estándar poblacional (σ) y n > 30, usamos el estadístico Z.
# Distribución bajo H0: Normal estándar Z.
# Fórmula del estadístico Z: Z = (x̄ - μ₀) / (σ / √n)
print("\n**3. Identificar el estadístico de prueba y su distribución**")
print("Dado que conocemos la desviación estándar poblacional (σ) y n > 30, usamos el estadístico Z.")
print("Distribución bajo H₀: Normal estándar Z.")
print("Fórmula del estadístico Z: Z = (x̄ - μ₀) / (σ / √n)")

# Datos proporcionados
n = 60       # Tamaño de la muestra
x_barra = 498 # Media muestral
sigma = 5    # Desviación estándar poblacional
mu_0 = 500   # Media hipotetizada (según H0)

# Calcular el error estándar
error_estandar = sigma / np.sqrt(n)

# 4. Determinar los valores críticos y la región de rechazo
# Para una prueba bilateral con α = 0.01, el α se divide en dos colas (α/2 = 0.005).
# Los valores críticos de Z son -Z(α/2) y +Z(α/2).
# norm.ppf(alpha/2) nos da el valor Z para la cola inferior.
# norm.ppf(1 - alpha/2) nos da el valor Z para la cola superior.
alpha_ci = alpha / 2
z_critico_inferior = norm.ppf(alpha_ci)
z_critico_superior = norm.ppf(1 - alpha_ci)

print(f"\n**4. Determinar los valores críticos y la región de rechazo**")
print(f"α/2 = {alpha_ci}")
print(f"Valores críticos de Z para α={alpha} (prueba bilateral): {z_critico_inferior:.4f} y {z_critico_superior:.4f}")
print(f"Región de rechazo: Z_calculado ≤ {z_critico_inferior:.4f} o Z_calculado ≥ {z_critico_superior:.4f}")


# 5. Calcular el estadístico de prueba y el valor p
# Calcular el estadístico Z
z_estadistico = (x_barra - mu_0) / error_estandar
# Calcular el valor p para una prueba bilateral: 2 * P(Z <= -|z_estadistico|) o 2 * P(Z >= |z_estadistico|)
# Usamos 2 * norm.cdf(-abs(z_estadistico)) para obtener el valor p correcto para prueba bilateral
valor_p = 2 * norm.cdf(-abs(z_estadistico))

print(f"\n**5. Calcular el estadístico de prueba y el valor p**")
print(f"Estadístico Z calculado: {z_estadistico:.4f}")
print(f"Valor p: {valor_p:.4f}")

# 6. Graficar la distribución, región de rechazo, valores críticos y estadístico calculado
print(f"\n**6. Graficar la distribución, región de rechazo, valores críticos y estadístico calculado**")
# Rango de valores para graficar la distribución normal estándar
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1) # PDF de la distribución normal estándar (media=0, desviación estándar=1)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución Normal Estándar')

# Rellenar la región de rechazo (colas izquierda y derecha)
x_reject_inferior = np.linspace(-4, z_critico_inferior, 100)
y_reject_inferior = norm.pdf(x_reject_inferior, 0, 1)
plt.fill_between(x_reject_inferior, y_reject_inferior, color='red', alpha=0.5, label=f'Región de Rechazo Inferior (α/2={alpha_ci})')

x_reject_superior = np.linspace(z_critico_superior, 4, 100)
y_reject_superior = norm.pdf(x_reject_superior, 0, 1)
plt.fill_between(x_reject_superior, y_reject_superior, color='red', alpha=0.5, label=f'Región de Rechazo Superior (α/2={alpha_ci})')

# Marcar los valores críticos de Z
plt.axvline(z_critico_inferior, color='red', linestyle='--', label=f'Valor Crítico Z (Inf) = {z_critico_inferior:.4f}')
plt.axvline(z_critico_superior, color='red', linestyle='--', label=f'Valor Crítico Z (Sup) = {z_critico_superior:.4f}')

# Marcar el estadístico Z calculado
plt.axvline(z_estadistico, color='blue', linestyle='-', label=f'Estadístico Z Calculado = {z_estadistico:.4f}')

# Añadir etiquetas y título
plt.title('Distribución Normal Estándar con Región de Rechazo y Estadístico Z (Prueba Bilateral)')
plt.xlabel('Valor de Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Tomar una decisión estadística y concluir en términos prácticos
print(f"\n**7. Tomar una decisión estadística y concluir en términos prácticos**")
print("Para tomar la decisión, comparamos el estadístico Z calculado con los valores críticos de Z, o el valor p con el nivel de significancia α.")

print(f"\nComparando estadístico Z vs. valores críticos:")
print(f"- Estadístico Z calculado: {z_estadistico:.4f}")
print(f"- Valores críticos Z: {z_critico_inferior:.4f} y {z_critico_superior:.4f}")
if z_estadistico <= z_critico_inferior or z_estadistico >= z_critico_superior:
    print(f"Como |{z_estadistico:.4f}| ≥ |{z_critico_superior:.4f}| (ya que {abs(z_estadistico):.4f} ≥ {abs(z_critico_superior):.4f}), el estadístico Z calculado cae en la región de rechazo. **Rechazamos la hipótesis nula (H₀)**.")
else:
    print(f"Como |{z_estadistico:.4f}| < |{z_critico_superior:.4f}| (ya que {abs(z_estadistico):.4f} < {abs(z_critico_superior):.4f}), el estadístico Z calculado NO cae en la región de rechazo. **No rechazamos la hipótesis nula (H₀)**.")

print(f"\nComparando valor p vs. α:")
print(f"- Valor p: {valor_p:.4f}")
print(f"- α: {alpha}")
if valor_p <= alpha:
    print(f"Como {valor_p:.4f} ≤ {alpha}, el valor p es menor que α. **Rechazamos la hipótesis nula (H₀)**.")
else:
     print(f"Como {valor_p:.4f} > {alpha}, el valor p es mayor que α. **No rechazamos la hipótesis nula (H₀)**.")

print("\n**Conclusión en términos prácticos:**")
if valor_p <= alpha: # Usar valor p para la conclusión general para mayor claridad
    print(f"Con un nivel de significancia del {alpha*100}%, hay suficiente evidencia estadística para concluir que la media de llenado de las botellas es significativamente diferente a 500 mL.")
else:
    print(f"Con un nivel de significancia del {alpha*100}%, no hay suficiente evidencia estadística para concluir que la media de llenado de las botellas es significativamente diferente a 500 mL.")

print("\n**Significado del valor p:**")
print(f"El valor p de {valor_p:.4f} significa que, si la media real de llenado fuera de 500 mL (es decir, si H₀ fuera verdadera), solo habría un {valor_p*100:.2f}% de probabilidad de observar una media muestral de {x_barra} mL o algo más extremo (ya sea mayor o menor que 500) simplemente por casualidad en una muestra de {n} botellas.")
if valor_p <= alpha:
     print(f"Dado que esta probabilidad es baja (menor que nuestro umbral α de {alpha}), consideramos que el resultado observado es poco probable bajo H₀, lo que nos lleva a rechazarla.")
else:
    print(f"Dado que esta probabilidad es alta (mayor que nuestro umbral α de {alpha}), consideramos que el resultado observado es probable bajo H₀, lo que nos lleva a no rechazarla.")
```

```python
import numpy as np
from scipy.stats import t
import matplotlib.pyplot as plt

# Problema: Prueba de Hipótesis para la Media (σ desconocida, n < 30) - Promedio en matemáticas
# Contexto: Un colegio afirma que el promedio de su clase en matemáticas es 80/100 (μ0).
# Una muestra de 12 estudiantes tiene una media de 75 (x̄) y desviación estándar muestral s = 8.
# Con un nivel de significancia de α = 0,01.
# Objetivo: Determinar si hay suficiente evidencia estadística para concluir que el promedio de la clase es diferente a 80.

print("### Problema: Prueba de Hipótesis para la Media (σ desconocida, n < 30) - Promedio en matemáticas")
print("\n**Contexto:** Un colegio afirma que el promedio de su clase en matemáticas es 80/100 (μ₀). Una muestra de 12 estudiantes tiene una media de 75 (x̄) y desviación estándar muestral s = 8. Con un nivel de significancia de α = 0.01.")
print("\n**Objetivo:** Determinar si hay suficiente evidencia estadística para concluir que el promedio de la clase es diferente a 80.")

# 1. Plantear las hipótesis nula y alternativa
# H0: μ = 80 (El promedio de la clase es de 80)
# H1: μ ≠ 80 (El promedio de la clase es diferente a 80)
# Esta es una prueba de hipótesis bilateral (de dos colas).
print("\n**1. Plantear las hipótesis nula y alternativa**")
print("H₀: μ = 80 (El promedio de la clase es de 80)")
print("H₁: μ ≠ 80 (El promedio de la clase es diferente a 80)")
print("Esta es una prueba de hipótesis bilateral (de dos colas).")

# 2. Establecer el nivel de significancia α
alpha = 0.01
print(f"\n**2. Establecer el nivel de significancia α**")
print(f"α = {alpha}")

# 3. Identificar el estadístico de prueba y su distribución
# Dado que NO conocemos la desviación estándar poblacional (σ) y n < 30, usamos el estadístico t de Student.
# Distribución bajo H0: Distribución t de Student con n-1 grados de libertad.
# Fórmula del estadístico t: t = (x̄ - μ₀) / (s / √n)
print("\n**3. Identificar el estadístico de prueba y su distribución**")
print("Dado que NO conocemos la desviación estándar poblacional (σ) y n < 30, usamos el estadístico t de Student.")
print("Distribución bajo H₀: Distribución t de Student con n-1 grados de libertad.")
print("Fórmula del estadístico t: t = (x̄ - μ₀) / (s / √n)")

# Datos proporcionados
n = 12       # Tamaño de la muestra
x_barra = 75 # Media muestral
s = 8        # Desviación estándar muestral
mu_0 = 80    # Media hipotetizada (según H0)
grados_libertad = n - 1 # Grados de libertad

# Calcular el error estándar muestral
error_estandar = s / np.sqrt(n)

# 4. Determinar los grados de libertad, los valores críticos y la región de rechazo
# Para una prueba bilateral con α = 0.01 y n-1 grados de libertad,
# el α se divide en dos colas (α/2 = 0.005).
# Los valores críticos de t son -t(α/2) y +t(α/2) con n-1 grados de libertad.
alpha_ci = alpha / 2
t_critico_inferior = t.ppf(alpha_ci, grados_libertad)
t_critico_superior = t.ppf(1 - alpha_ci, grados_libertad)


print(f"\n**4. Determinar los grados de libertad, los valores críticos y la región de rechazo**")
print(f"Grados de libertad (df): {grados_libertad}")
print(f"α/2 = {alpha_ci}")
print(f"Valores críticos de t para α={alpha} (prueba bilateral): {t_critico_inferior:.4f} y {t_critico_superior:.4f}")
print(f"Región de rechazo: t_calculado ≤ {t_critico_inferior:.4f} o t_calculado ≥ {t_critico_superior:.4f}")


# 5. Calcular el estadístico de prueba y el valor p
# Calcular el estadístico t
t_estadistico = (x_barra - mu_0) / error_estandar
# Calcular el valor p para una prueba bilateral: 2 * P(T <= -|t_estadistico|) o 2 * P(T >= |t_estadistico|)
# Usamos 2 * t.cdf(-abs(t_estadistico), grados_libertad) para obtener el valor p correcto para prueba bilateral
valor_p = 2 * t.cdf(-abs(t_estadistico), grados_libertad)


print(f"\n**5. Calcular el estadístico de prueba y el valor p**")
print(f"Estadístico t calculado: {t_estadistico:.4f}")
print(f"Valor p: {valor_p:.4f}")

# 6. Graficar la distribución, región de rechazo, valores críticos y estadístico calculado
print(f"\n**6. Graficar la distribución, región de rechazo, valores críticos y estadístico calculado**")
# Rango de valores para graficar la distribución t
x = np.linspace(t.ppf(0.0001, grados_libertad), t.ppf(0.9999, grados_libertad), 1000) # Ampliar rango para mejor visualización
y = t.pdf(x, grados_libertad) # PDF de la distribución t

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'Distribución t (df={grados_libertad})')

# Rellenar la región de rechazo (colas izquierda y derecha)
x_reject_inferior = np.linspace(t.ppf(0.0001, grados_libertad), t_critico_inferior, 100)
y_reject_inferior = t.pdf(x_reject_inferior, grados_libertad)
plt.fill_between(x_reject_inferior, y_reject_inferior, color='red', alpha=0.5, label=f'Región de Rechazo Inferior (α/2={alpha_ci})')

x_reject_superior = np.linspace(t_critico_superior, t.ppf(0.9999, grados_libertad), 100)
y_reject_superior = t.pdf(x_reject_superior, grados_libertad)
plt.fill_between(x_reject_superior, y_reject_superior, color='red', alpha=0.5, label=f'Región de Rechazo Superior (α/2={alpha_ci})')

# Marcar los valores críticos de t
plt.axvline(t_critico_inferior, color='red', linestyle='--', label=f'Valor Crítico t (Inf) = {t_critico_inferior:.4f}')
plt.axvline(t_critico_superior, color='red', linestyle='--', label=f'Valor Crítico t (Sup) = {t_critico_superior:.4f}')

# Marcar el estadístico t calculado
plt.axvline(t_estadistico, color='blue', linestyle='-', label=f'Estadístico t Calculado = {t_estadistico:.4f}')

# Añadir etiquetas y título
plt.title(f'Distribución t (df={grados_libertad}) con Región de Rechazo y Estadístico t (Prueba Bilateral)')
plt.xlabel('Valor de t')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Tomar una decisión estadística y concluir en términos prácticos
print(f"\n**7. Tomar una decisión estadística y concluir en términos prácticos**")
print("Para tomar la decisión, comparamos el estadístico t calculado con los valores críticos de t, o el valor p con el nivel de significancia α.")

print(f"\nComparando estadístico t vs. valores críticos:")
print(f"- Estadístico t calculado: {t_estadistico:.4f}")
print(f"- Valores críticos t: {t_critico_inferior:.4f} y {t_critico_superior:.4f}")
if t_estadistico <= t_critico_inferior or t_estadistico >= t_critico_superior:
    print(f"Como |{t_estadistico:.4f}| ≥ |{t_critico_superior:.4f}| (ya que {abs(t_estadistico):.4f} ≥ {abs(t_critico_superior):.4f}), el estadístico t calculado cae en la región de rechazo. **Rechazamos la hipótesis nula (H₀)**.")
else:
    print(f"Como |{t_estadistico:.4f}| < |{t_critico_superior:.4f}| (ya que {abs(t_estadistico):.4f} < {abs(t_critico_superior):.4f}), el estadístico t calculado NO cae en la región de rechazo. **No rechazamos la hipótesis nula (H₀)**.")

print(f"\nComparando valor p vs. α:")
print(f"- Valor p: {valor_p:.4f}")
print(f"- α: {alpha}")
if valor_p <= alpha:
    print(f"Como {valor_p:.4f} ≤ {alpha}, el valor p es menor que α. **Rechazamos la hipótesis nula (H₀)**.")
else:
     print(f"Como {valor_p:.4f} > {alpha}, el valor p es mayor que α. **No rechazamos la hipótesis nula (H₀)**.")

print("\n**Conclusión en términos prácticos:**")
if valor_p <= alpha: # Usar valor p para la conclusión general para mayor claridad
    print(f"Con un nivel de significancia del {alpha*100}%, hay suficiente evidencia estadística para concluir que el promedio de la clase de matemáticas es significativamente diferente a 80/100.")
else:
    print(f"Con un nivel de significancia del {alpha*100}%, no hay suficiente evidencia estadística para concluir que el promedio de la clase de matemáticas es significativamente diferente a 80/100.")

print("\n**Significado del valor p:**")
print(f"El valor p de {valor_p:.4f} significa que, si el promedio real de la clase fuera de 80 (es decir, si H₀ fuera verdadera), solo habría un {valor_p*100:.2f}% de probabilidad de observar una media muestral de {x_barra} o algo más extremo (ya sea mayor o menor que 80) simplemente por casualidad en una muestra de {n} estudiantes.")
if valor_p <= alpha:
     print(f"Dado que esta probabilidad es baja (menor que nuestro umbral α de {alpha}), consideramos que el resultado observado es poco probable bajo H₀, lo que nos lleva a rechazarla.")
else:
    print(f"Dado que esta probabilidad es alta (mayor que nuestro umbral α de {alpha}), consideramos que el resultado observado es probable bajo H₀, lo que nos lleva a no rechazarla.")
```

```python
import numpy as np
from scipy.stats import t
import matplotlib.pyplot as plt

# Problema: Prueba de Hipótesis para la Media (σ desconocida, n < 30) - Nivel de hierro
# Contexto: Un suplemento promete aumentar el nivel de hierro en la sangre en 2 mg/dL en promedio (μ0).
# En 8 pacientes, el aumento promedio fue 1.5 mg/dL (x̄) con desviación estándar muestral s = 0.6 mg/dL.
# Con un nivel de significancia de α = 0,10.
# Objetivo: Determinar si hay suficiente evidencia estadística para concluir que el aumento promedio es diferente a 2 mg/dL.
# NOTA: La afirmación es que aumenta "en 2 mg/dL". La muestra muestra un aumento menor.
# Interpretaremos como una prueba bilateral: ¿el aumento promedio es *diferente* a 2 mg/dL?

print("### Problema: Prueba de Hipótesis para la Media (σ desconocida, n < 30) - Nivel de hierro")
print("\n**Contexto:** Un suplemento promete aumentar el nivel de hierro en la sangre en 2 mg/dL en promedio (μ₀). En 8 pacientes, el aumento promedio fue 1.5 mg/dL (x̄) con desviación estándar muestral s = 0.6 mg/dL. Con un nivel de significancia de α = 0.10.")
print("\n**Objetivo:** Determinar si hay suficiente evidencia estadística para concluir que el aumento promedio es diferente a 2 mg/dL.")
print("\n*Nota: Interpretamos la afirmación como una prueba bilateral para ver si el aumento es significativamente diferente de 2 mg/dL.*")

# 1. Plantear las hipótesis nula y alternativa
# H0: μ = 2 (El aumento promedio es de 2 mg/dL)
# H1: μ ≠ 2 (El aumento promedio es diferente a 2 mg/dL)
# Esta es una prueba de hipótesis bilateral (de dos colas).
print("\n**1. Plantear las hipótesis nula y alternativa**")
print("H₀: μ = 2 (El aumento promedio es de 2 mg/dL)")
print("H₁: μ ≠ 2 (El aumento promedio es diferente a 2 mg/dL)")
print("Esta es una prueba de hipótesis bilateral (de dos colas).")

# 2. Establecer el nivel de significancia α
alpha = 0.10
print(f"\n**2. Establecer el nivel de significancia α**")
print(f"α = {alpha}")

# 3. Identificar el estadístico de prueba y su distribución
# Dado que NO conocemos la desviación estándar poblacional (σ) y n < 30, usamos el estadístico t de Student.
# Distribución bajo H0: Distribución t de Student con n-1 grados de libertad.
# Fórmula del estadístico t: t = (x̄ - μ₀) / (s / √n)
print("\n**3. Identificar el estadístico de prueba y su distribución**")
print("Dado que NO conocemos la desviación estándar poblacional (σ) y n < 30, usamos el estadístico t de Student.")
print("Distribución bajo H₀: Distribución t de Student con n-1 grados de libertad.")
print("Fórmula del estadístico t: t = (x̄ - μ₀) / (s / √n)")

# Datos proporcionados
n = 8       # Tamaño de la muestra
x_barra = 1.5 # Media muestral
s = 0.6      # Desviación estándar muestral
mu_0 = 2    # Media hipotetizada (según H0)
grados_libertad = n - 1 # Grados de libertad

# Calcular el error estándar muestral
error_estandar = s / np.sqrt(n)

# 4. Determinar los grados de libertad, los valores críticos y la región de rechazo
# Para una prueba bilateral con α = 0.10 y n-1 grados de libertad,
# el α se divide en dos colas (α/2 = 0.05).
# Los valores críticos de t son -t(α/2) y +t(α/2) con n-1 grados de libertad.
alpha_ci = alpha / 2
t_critico_inferior = t.ppf(alpha_ci, grados_libertad)
t_critico_superior = t.ppf(1 - alpha_ci, grados_libertad)


print(f"\n**4. Determinar los grados de libertad, los valores críticos y la región de rechazo**")
print(f"Grados de libertad (df): {grados_libertad}")
print(f"α/2 = {alpha_ci}")
print(f"Valores críticos de t para α={alpha} (prueba bilateral): {t_critico_inferior:.4f} y {t_critico_superior:.4f}")
print(f"Región de rechazo: t_calculado ≤ {t_critico_inferior:.4f} o t_calculado ≥ {t_critico_superior:.4f}")


# 5. Calcular el estadístico de prueba y el valor p
# Calcular el estadístico t
t_estadistico = (x_barra - mu_0) / error_estandar
# Calcular el valor p para una prueba bilateral: 2 * P(T <= -|t_estadistico|) o 2 * P(T >= |t_estadistico|)
# Usamos 2 * t.cdf(-abs(t_estadistico), grados_libertad) para obtener el valor p correcto para prueba bilateral
valor_p = 2 * t.cdf(-abs(t_estadistico), grados_libertad)


print(f"\n**5. Calcular el estadístico de prueba y el valor p**")
print(f"Estadístico t calculado: {t_estadistico:.4f}")
print(f"Valor p: {valor_p:.4f}")

# 6. Graficar la distribución, región de rechazo, valores críticos y estadístico calculado
print(f"\n**6. Graficar la distribución, región de rechazo, valores críticos y estadístico calculado**")
# Rango de valores para graficar la distribución t
x = np.linspace(t.ppf(0.001, grados_libertad), t.ppf(0.999, grados_libertad), 1000)
y = t.pdf(x, grados_libertad) # PDF de la distribución t

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'Distribución t (df={grados_libertad})')

# Rellenar la región de rechazo (colas izquierda y derecha)
x_reject_inferior = np.linspace(t.ppf(0.001, grados_libertad), t_critico_inferior, 100)
y_reject_inferior = t.pdf(x_reject_inferior, grados_libertad)
plt.fill_between(x_reject_inferior, y_reject_inferior, color='red', alpha=0.5, label=f'Región de Rechazo Inferior (α/2={alpha_ci})')

x_reject_superior = np.linspace(t_critico_superior, t.ppf(0.999, grados_libertad), 100)
y_reject_superior = t.pdf(x_reject_superior, grados_libertad)
plt.fill_between(x_reject_superior, y_reject_superior, color='red', alpha=0.5, label=f'Región de Rechazo Superior (α/2={alpha_ci})')

# Marcar los valores críticos de t
plt.axvline(t_critico_inferior, color='red', linestyle='--', label=f'Valor Crítico t (Inf) = {t_critico_inferior:.4f}')
plt.axvline(t_critico_superior, color='red', linestyle='--', label=f'Valor Crítico t (Sup) = {t_critico_superior:.4f}')

# Marcar el estadístico t calculado
plt.axvline(t_estadistico, color='blue', linestyle='-', label=f'Estadístico t Calculado = {t_estadistico:.4f}')

# Añadir etiquetas y título
plt.title(f'Distribución t (df={grados_libertad}) con Región de Rechazo y Estadístico t (Prueba Bilateral)')
plt.xlabel('Valor de t')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Tomar una decisión estadística y concluir en términos prácticos
print(f"\n**7. Tomar una decisión estadística y concluir en términos prácticos**")
print("Para tomar la decisión, comparamos el estadístico t calculado con los valores críticos de t, o el valor p con el nivel de significancia α.")

print(f"\nComparando estadístico t vs. valores críticos:")
print(f"- Estadístico t calculado: {t_estadistico:.4f}")
print(f"- Valores críticos t: {t_critico_inferior:.4f} y {t_critico_superior:.4f}")
if t_estadistico <= t_critico_inferior or t_estadistico >= t_critico_superior:
    print(f"Como |{t_estadistico:.4f}| ≥ |{t_critico_superior:.4f}| (ya que {abs(t_estadistico):.4f} ≥ {abs(t_critico_superior):.4f}), el estadístico t calculado cae en la región de rechazo. **Rechazamos la hipótesis nula (H₀)**.")
else:
    print(f"Como |{t_estadistico:.4f}| < |{t_critico_superior:.4f}| (ya que {abs(t_estadistico):.4f} < {abs(t_critico_superior):.4f}), el estadístico t calculado NO cae en la región de rechazo. **No rechazamos la hipótesis nula (H₀)**.")

print(f"\nComparando valor p vs. α:")
print(f"- Valor p: {valor_p:.4f}")
print(f"- α: {alpha}")
if valor_p <= alpha:
    print(f"Como {valor_p:.4f} ≤ {alpha}, el valor p es menor que α. **Rechazamos la hipótesis nula (H₀)**.")
else:
     print(f"Como {valor_p:.4f} > {alpha}, el valor p es mayor que α. **No rechazamos la hipótesis nula (H₀)**.")

print("\n**Conclusión en términos prácticos:**")
if valor_p <= alpha: # Usar valor p para la conclusión general para mayor claridad
    print(f"Con un nivel de significancia del {alpha*100}%, hay suficiente evidencia estadística para concluir que el aumento promedio de hierro en la sangre con el suplemento es significativamente diferente a 2 mg/dL.")
else:
    print(f"Con un nivel de significancia del {alpha*100}%, no hay suficiente evidencia estadística para concluir que el aumento promedio de hierro en la sangre con el suplemento es significativamente diferente a 2 mg/dL.")

print("\n**Significado del valor p:**")
print(f"El valor p de {valor_p:.4f} significa que, si el aumento promedio real fuera de 2 mg/dL (es decir, si H₀ fuera verdadera), solo habría un {valor_p*100:.2f}% de probabilidad de observar una media muestral de {x_barra} mg/dL o algo más extremo (ya sea mayor o menor que 2) simplemente por casualidad en una muestra de {n} pacientes.")
if valor_p <= alpha:
     print(f"Dado que esta probabilidad es baja (menor que nuestro umbral α de {alpha}), consideramos que el resultado observado es poco probable bajo H₀, lo que nos lleva a rechazarla.")
else:
    print(f"Dado que esta probabilidad es alta (mayor que nuestro umbral α de {alpha}), consideramos que el resultado observado es probable bajo H₀, lo que nos lleva a no rechazarla.")
```

```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt

# Problema: Prueba de Hipótesis para la Varianza (Distribución Chi-cuadrada)
# Contexto: Una compañía de telefonía afirma que la varianza del tiempo de duración de las llamadas es de σ² = 2.5 minutos².
# Se toma una muestra de n = 16 llamadas y se obtiene una varianza muestral de s² = 4.
# ¿Hay evidencia al 5 % de significancia de que la varianza poblacional difiere de 2.5?
# Objetivo: Determinar si hay suficiente evidencia estadística para concluir que la varianza poblacional difiere de 2.5.

print("### Problema: Prueba de Hipótesis para la Varianza (Distribución Chi-cuadrada)")
print("\n**Contexto:** Una compañía de telefonía afirma que la varianza del tiempo de duración de las llamadas es de σ² = 2.5 minutos². Se toma una muestra de n = 16 llamadas y se obtiene una varianza muestral de s² = 4. ¿Hay evidencia al 5 % de significancia de que la varianza poblacional difiere de 2.5?")
print("\n**Objetivo:** Determinar si hay suficiente evidencia estadística para concluir que la varianza poblacional difiere de 2.5.")

# 1. Plantear las hipótesis nula y alternativa
# H0: σ² = 2.5 (La varianza poblacional es 2.5)
# H1: σ² ≠ 2.5 (La varianza poblacional difiere de 2.5)
# Esta es una prueba de hipótesis bilateral (de dos colas).
print("\n**1. Plantear las hipótesis nula y alternativa**")
print("H₀: σ² = 2.5 (La varianza poblacional es 2.5)")
print("H₁: σ² ≠ 2.5 (La varianza poblacional difiere de 2.5)")
print("Esta es una prueba de hipótesis bilateral (de dos colas).")

# 2. Establecer el nivel de significancia α
alpha = 0.05
print(f"\n**2. Establecer el nivel de significancia α**")
print(f"α = {alpha}")

# 3. Identificar el estadístico de prueba y su distribución
# Para probar una hipótesis sobre la varianza poblacional (σ²), usamos el estadístico Chi-cuadrada (χ²).
# Distribución bajo H0: Distribución Chi-cuadrada con n-1 grados de libertad.
# Fórmula del estadístico χ²: χ² = ((n - 1) * s²) / σ²
print("\n**3. Identificar el estadístico de prueba y su distribución**")
print("Para probar una hipótesis sobre la varianza poblacional (σ²), usamos el estadístico Chi-cuadrada (χ²).")
print("Distribución bajo H₀: Distribución Chi-cuadrada con n-1 grados de libertad.")
print("Fórmula del estadístico χ²: χ² = ((n - 1) * s²) / σ²")

# Datos proporcionados
n = 16       # Tamaño de la muestra
s2 = 4       # Varianza muestral
sigma2_0 = 2.5 # Varianza hipotetizada (según H0)
grados_libertad = n - 1 # Grados de libertad

# 4. Determine los grados de libertad, el valor crítico, la región crítica y la región de rechazo.
# Para una prueba bilateral con α = 0.05 y n-1 grados de libertad,
# el α se divide en dos colas (α/2 = 0.025).
# Los valores críticos de χ² son χ²(α/2, gl) y χ²(1 - α/2, gl).
alpha_ci = alpha / 2
chi2_critico_inferior = chi2.ppf(alpha_ci, grados_libertad)
chi2_critico_superior = chi2.ppf(1 - alpha_ci, grados_libertad)

print(f"\n**4. Determine los grados de libertad, el valor crítico, la región crítica y la región de rechazo.**")
print(f"Grados de libertad (df): {grados_libertad}")
print(f"α/2 = {alpha_ci}")
print(f"Valores críticos de χ² para α={alpha} (prueba bilateral): {chi2_critico_inferior:.4f} y {chi2_critico_superior:.4f}")
print(f"Región de rechazo: χ²_calculado ≤ {chi2_critico_inferior:.4f} o χ²_calculado ≥ {chi2_critico_superior:.4f}")


# 5. Calcule el estadístico χ² y el valor p.
# Calcular el estadístico χ²
chi2_estadistico = ((n - 1) * s2) / sigma2_0
# Calcular el valor p para una prueba bilateral
# El valor p es la suma de las áreas en las dos colas que están más allá del estadístico calculado.
# Si el estadístico calculado está a la derecha del centro, calculamos el área a la derecha y la multiplicamos por 2.
# Si el estadístico calculado está a la izquierda del centro, calculamos el área a la izquierda y la multiplicamos por 2.
# O más generalmente: p-value = 2 * min(P(χ² <= χ²_estadistico), P(χ² >= χ²_estadistico))
# P(χ² <= χ²_estadistico) = chi2.cdf(chi2_estadistico, grados_libertad)
# P(χ² >= χ²_estadistico) = 1 - chi2.cdf(chi2_estadistico, grados_libertad)

valor_p_cola_izquierda = chi2.cdf(chi2_estadistico, grados_libertad)
valor_p_cola_derecha = 1 - chi2.cdf(chi2_estadistico, grados_libertad)
valor_p = 2 * min(valor_p_cola_izquierda, valor_p_cola_derecha)


print(f"\n**5. Calcule el estadístico χ² y el valor p.**")
print(f"Estadístico χ² calculado: {chi2_estadistico:.4f}")
print(f"Valor p: {valor_p:.4f}")


# 6. Graficar la distribución Chi-cuadrada, región de rechazo, valores críticos y estadístico calculado
print(f"\n**6. Graficar la distribución correspondiente, indicando claramente la región de rechazo, el valor crítico y el estadístico calculado.**")
# Rango de valores para graficar la distribución Chi-cuadrada
# Usamos los valores críticos para definir un rango apropiado para la gráfica
x = np.linspace(chi2.ppf(0.001, grados_libertad), chi2.ppf(0.999, grados_libertad), 1000)
y = chi2.pdf(x, grados_libertad) # PDF de la distribución Chi-cuadrada

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'Distribución χ² (df={grados_libertad})')

# Rellenar la región de rechazo (colas izquierda y derecha)
x_reject_inferior = np.linspace(chi2.ppf(0.001, grados_libertad), chi2_critico_inferior, 100)
y_reject_inferior = chi2.pdf(x_reject_inferior, grados_libertad)
plt.fill_between(x_reject_inferior, y_reject_inferior, color='red', alpha=0.5, label=f'Región de Rechazo Inferior (α/2={alpha_ci})')

x_reject_superior = np.linspace(chi2_critico_superior, chi2.ppf(0.999, grados_libertad), 100)
y_reject_superior = chi2.pdf(x_reject_superior, grados_libertad)
plt.fill_between(x_reject_superior, y_reject_superior, color='red', alpha=0.5, label=f'Región de Rechazo Superior (α/2={alpha_ci})')

# Marcar los valores críticos de χ²
plt.axvline(chi2_critico_inferior, color='red', linestyle='--', label=f'Valor Crítico χ² (Inf) = {chi2_critico_inferior:.4f}')
plt.axvline(chi2_critico_superior, color='red', linestyle='--', label=f'Valor Crítico χ² (Sup) = {chi2_critico_superior:.4f}')

# Marcar el estadístico χ² calculado
plt.axvline(chi2_estadistico, color='blue', linestyle='-', label=f'Estadístico χ² Calculado = {chi2_estadistico:.4f}')

# Añadir etiquetas y título
plt.title(f'Distribución χ² (df={grados_libertad}) con Región de Rechazo y Estadístico χ² (Prueba Bilateral)')
plt.xlabel('Valor de χ²')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Tomar una decisión estadística y concluir en términos prácticos
print(f"\n**7. Tomar una decisión estadística y concluir en términos prácticos**")
print("Para tomar la decisión, comparamos el estadístico χ² calculado con los valores críticos de χ², o el valor p con el nivel de significancia α.")

print(f"\nComparando estadístico χ² vs. valores críticos:")
print(f"- Estadístico χ² calculado: {chi2_estadistico:.4f}")
print(f"- Valores críticos χ²: {chi2_critico_inferior:.4f} y {chi2_critico_superior:.4f}")
if chi2_estadistico <= chi2_critico_inferior or chi2_estadistico >= chi2_critico_superior:
    print(f"Como {chi2_estadistico:.4f} es mayor que {chi2_critico_superior:.4f}, el estadístico χ² calculado cae en la región de rechazo. **Rechazamos la hipótesis nula (H₀)**.")
else:
    print(f"Como {chi2_estadistico:.4f} está entre {chi2_critico_inferior:.4f} y {chi2_critico_superior:.4f}, el estadístico χ² calculado NO cae en la región de rechazo. **No rechazamos la hipótesis nula (H₀)**.")

print(f"\nComparando valor p vs. α:")
print(f"- Valor p: {valor_p:.4f}")
print(f"- α: {alpha}")
if valor_p <= alpha:
    print(f"Como {valor_p:.4f} ≤ {alpha}, el valor p es menor que α. **Rechazamos la hipótesis nula (H₀)**.")
else:
     print(f"Como {valor_p:.4f} > {alpha}, el valor p es mayor que α. **No rechazamos la hipótesis nula (H₀)**.")

print("\n**Conclusión en términos prácticos:**")
if valor_p <= alpha: # Usar valor p para la conclusión general para mayor claridad
    print(f"Con un nivel de significancia del {alpha*100}%, hay suficiente evidencia estadística para concluir que la varianza poblacional del tiempo de duración de las llamadas es significativamente diferente a 2.5 minutos².")
else:
    print(f"Con un nivel de significancia del {alpha*100}%, no hay suficiente evidencia estadística para concluir que la varianza poblacional del tiempo de duración de las llamadas es significativamente diferente a 2.5 minutos².")

print("\n**Significado del valor p:**")
print(f"El valor p de {valor_p:.4f} significa que, si la varianza poblacional real fuera de 2.5 minutos² (es decir, si H₀ fuera verdadera), solo habría un {valor_p*100:.2f}% de probabilidad de observar una varianza muestral de {s2} o algo más extremo (ya sea mayor o menor que 2.5) simplemente por casualidad en una muestra de {n} llamadas.")
if valor_p <= alpha:
     print(f"Dado que esta probabilidad es baja (menor que nuestro umbral α de {alpha}), consideramos que el resultado observado es poco probable bajo H₀, lo que nos lleva a rechazarla.")
else:
    print(f"Dado que esta probabilidad es alta (mayor que nuestro umbral α de {alpha}), consideramos que el resultado observado es probable bajo H₀, lo que nos lleva a no rechazarla.")
```

```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt

# Problema: Prueba de Hipótesis para la Varianza (Distribución Chi-cuadrada) - Llenado de botellas
# Contexto: Una máquina embotelladora debe mantener la varianza del llenado en σ² = 1.2 ml².
# Se toma una muestra de n = 20 botellas y se encuentra s² = 2.0.
# ¿Existe evidencia al 1 % de que la varianza es mayor?
# Objetivo: Determinar si hay suficiente evidencia estadística para concluir que la varianza poblacional es mayor a 1.2 ml².

print("### Problema: Prueba de Hipótesis para la Varianza (Distribución Chi-cuadrada) - Llenado de botellas")
print("\n**Contexto:** Una máquina embotelladora debe mantener la varianza del llenado en σ² = 1.2 ml². Se toma una muestra de n = 20 botellas y se encuentra s² = 2.0. ¿Existe evidencia al 1 % de que la varianza es mayor?")
print("\n**Objetivo:** Determinar si hay suficiente evidencia estadística para concluir que la varianza poblacional es mayor a 1.2 ml².")

# 1. Plantear las hipótesis nula y alternativa
# H0: σ² = 1.2 (La varianza poblacional es 1.2)
# H1: σ² > 1.2 (La varianza poblacional es mayor a 1.2)
# Esta es una prueba de hipótesis unilateral (cola derecha).
print("\n**1. Plantear las hipótesis nula y alternativa**")
print("H₀: σ² = 1.2 (La varianza poblacional es 1.2)")
print("H₁: σ² > 1.2 (La varianza poblacional es mayor a 1.2)")
print("Esta es una prueba de hipótesis unilateral (cola derecha).")

# 2. Establecer el nivel de significancia α
alpha = 0.01
print(f"\n**2. Establecer el nivel de significancia α**")
print(f"α = {alpha}")

# 3. Identificar el estadístico de prueba y su distribución
# Para probar una hipótesis sobre la varianza poblacional (σ²), usamos el estadístico Chi-cuadrada (χ²).
# Distribución bajo H0: Distribución Chi-cuadrada con n-1 grados de libertad.
# Fórmula del estadístico χ²: χ² = ((n - 1) * s²) / σ²
print("\n**3. Identificar el estadístico de prueba y su distribución**")
print("Para probar una hipótesis sobre la varianza poblacional (σ²), usamos el estadístico Chi-cuadrada (χ²).")
print("Distribución bajo H₀: Distribución Chi-cuadrada con n-1 grados de libertad.")
print("Fórmula del estadístico χ²: χ² = ((n - 1) * s²) / σ²")

# Datos proporcionados
n = 20       # Tamaño de la muestra
s2 = 2.0     # Varianza muestral
sigma2_0 = 1.2 # Varianza hipotetizada (según H0)
grados_libertad = n - 1 # Grados de libertad

# 4. Determine los grados de libertad, el valor crítico, la región crítica y la región de rechazo.
# Para una prueba unilateral a la derecha con α = 0.01 y n-1 grados de libertad,
# el valor crítico de χ² deja el 1% del área en la cola derecha.
# chi2.ppf(1 - alpha, grados_libertad) nos da el valor χ² tal que P(χ² <= χ²_critico) = 1 - alpha
chi2_critico = chi2.ppf(1 - alpha, grados_libertad)

print(f"\n**4. Determine los grados de libertad, el valor crítico, la región crítica y la región de rechazo.**")
print(f"Grados de libertad (df): {grados_libertad}")
print(f"Valor crítico de χ² para α={alpha} (cola derecha): {chi2_critico:.4f}")
print(f"Región de rechazo: χ²_calculado ≥ {chi2_critico:.4f}")


# 5. Calcule el estadístico χ² y el valor p.
# Calcular el estadístico χ²
chi2_estadistico = ((n - 1) * s2) / sigma2_0
# Calcular el valor p para una prueba unilateral derecha: P(χ² >= χ²_estadistico)
# 1 - chi2.cdf(chi2_estadistico, grados_libertad) nos da la probabilidad en la cola derecha
valor_p = 1 - chi2.cdf(chi2_estadistico, grados_libertad)


print(f"\n**5. Calcule el estadístico χ² y el valor p.**")
print(f"Estadístico χ² calculado: {chi2_estadistico:.4f}")
print(f"Valor p: {valor_p:.4f}")


# 6. Graficar la distribución Chi-cuadrada, región de rechazo, valor crítico y estadístico calculado
print(f"\n**6. Graficar la distribución correspondiente, indicando claramente la región de rechazo, el valor crítico y el estadístico calculado.**")
# Rango de valores para graficar la distribución Chi-cuadrada
# Usamos los valores críticos y el estadístico calculado para definir un rango apropiado para la gráfica
max_val = max(chi2_critico, chi2_estadistico, chi2.ppf(0.999, grados_libertad)) * 1.1 # Asegurar que el rango sea suficiente
x = np.linspace(chi2.ppf(0.001, grados_libertad), max_val, 1000)
y = chi2.pdf(x, grados_libertad) # PDF de la distribución Chi-cuadrada

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'Distribución χ² (df={grados_libertad})')

# Rellenar la región de rechazo (cola derecha)
x_reject = np.linspace(chi2_critico, max_val, 100)
y_reject = chi2.pdf(x_reject, grados_libertad)
plt.fill_between(x_reject, y_reject, color='red', alpha=0.5, label=f'Región de Rechazo (α={alpha})')

# Marcar el valor crítico de χ²
plt.axvline(chi2_critico, color='red', linestyle='--', label=f'Valor Crítico χ² = {chi2_critico:.4f}')

# Marcar el estadístico χ² calculado
plt.axvline(chi2_estadistico, color='blue', linestyle='-', label=f'Estadístico χ² Calculado = {chi2_estadistico:.4f}')

# Añadir etiquetas y título
plt.title(f'Distribución χ² (df={grados_libertad}) con Región de Rechazo y Estadístico χ² (Prueba Unilateral Derecha)')
plt.xlabel('Valor de χ²')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Tomar una decisión estadística y concluir en términos prácticos
print(f"\n**7. Tomar una decisión estadística y concluir en términos prácticos**")
print("Para tomar la decisión, comparamos el estadístico χ² calculado con el valor crítico de χ², o el valor p con el nivel de significancia α.")

print(f"\nComparando estadístico χ² vs. valor crítico:")
print(f"- Estadístico χ² calculado: {chi2_estadistico:.4f}")
print(f"- Valor crítico χ²: {chi2_critico:.4f}")
if chi2_estadistico >= chi2_critico:
    print(f"Como {chi2_estadistico:.4f} ≥ {chi2_critico:.4f}, el estadístico χ² calculado cae en la región de rechazo. **Rechazamos la hipótesis nula (H₀)**.")
else:
    print(f"Como {chi2_estadistico:.4f} < {chi2_critico:.4f}, el estadístico χ² calculado NO cae en la región de rechazo. **No rechazamos la hipótesis nula (H₀)**.")

print(f"\nComparando valor p vs. α:")
print(f"- Valor p: {valor_p:.4f}")
print(f"- α: {alpha}")
if valor_p <= alpha:
    print(f"Como {valor_p:.4f} ≤ {alpha}, el valor p es menor que α. **Rechazamos la hipótesis nula (H₀)**.")
else:
     print(f"Como {valor_p:.4f} > {alpha}, el valor p es mayor que α. **No rechazamos la hipótesis nula (H₀)**.")

print("\n**Conclusión en términos prácticos:**")
if valor_p <= alpha: # Usar valor p para la conclusión general para mayor claridad
    print(f"Con un nivel de significancia del {alpha*100}%, hay suficiente evidencia estadística para concluir que la varianza poblacional del llenado de las botellas es significativamente mayor a 1.2 ml².")
else:
    print(f"Con un nivel de significancia del {alpha*100}%, no hay suficiente evidencia estadística para concluir que la varianza poblacional del llenado de las botellas es significativamente mayor a 1.2 ml².")

print("\n**Significado del valor p:**")
print(f"El valor p de {valor_p:.4f} significa que, si la varianza poblacional real fuera de 1.2 ml² (es decir, si H₀ fuera verdadera), solo habría un {valor_p*100:.2f}% de probabilidad de observar una varianza muestral de {s2} o algo más extremo (mayor que 1.2) simplemente por casualidad en una muestra de {n} botellas.")
if valor_p <= alpha:
     print(f"Dado que esta probabilidad es baja (menor que nuestro umbral α de {alpha}), consideramos que el resultado observado es poco probable bajo H₀, lo que nos lleva a rechazarla.")
else:
    print(f"Dado que esta probabilidad es alta (mayor que nuestro umbral α de {alpha}), consideramos que el resultado observado es probable bajo H₀, lo que nos lleva a no rechazarla.")
```

```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt

# Problema: Prueba de Hipótesis para la Varianza (Distribución Chi-cuadrada) - Grosor de piezas
# Contexto: El control de calidad de una fábrica sostiene que la varianza en el grosor de piezas metálicas es de al menos σ² = 0.5 mm².
# Una muestra de n = 12 piezas dio una varianza s² = 0.3.
# ¿Se puede concluir al 5 % de significancia que la varianza poblacional es menor que la garantizada?
# Objetivo: Determinar si hay suficiente evidencia estadística para concluir que la varianza poblacional es menor que 0.5 mm².

print("### Problema: Prueba de Hipótesis para la Varianza (Distribución Chi-cuadrada) - Grosor de piezas")
print("\n**Contexto:** El control de calidad de una fábrica sostiene que la varianza en el grosor de piezas metálicas es de al menos σ² = 0.5 mm². Una muestra de n = 12 piezas dio una varianza s² = 0.3. ¿Se puede concluir al 5 % de significancia que la varianza poblacional es menor que la garantizada?")
print("\n**Objetivo:** Determinar si hay suficiente evidencia estadística para concluir que la varianza poblacional es menor que 0.5 mm².")


# 1. Plantear las hipótesis nula y alternativa
# H0: σ² = 0.5 (La varianza poblacional es 0.5)
# H1: σ² < 0.5 (La varianza poblacional es menor a 0.5 - menor que la garantizada)
# Esta es una prueba de hipótesis unilateral (cola izquierda).
print("\n**1. Plantear las hipótesis nula y alternativa**")
print("H₀: σ² = 0.5 (La varianza poblacional es 0.5)")
print("H₁: σ² < 0.5 (La varianza poblacional es menor a 0.5 - menor que la garantizada)")
print("Esta es una prueba de hipótesis unilateral (cola izquierda).")

# 2. Establecer el nivel de significancia α
alpha = 0.05
print(f"\n**2. Establecer el nivel de significancia α**")
print(f"α = {alpha}")

# 3. Identificar el estadístico de prueba y su distribución
# Para probar una hipótesis sobre la varianza poblacional (σ²), usamos el estadístico Chi-cuadrada (χ²).
# Distribución bajo H0: Distribución Chi-cuadrada con n-1 grados de libertad.
# Fórmula del estadístico χ²: χ² = ((n - 1) * s²) / σ²
print("\n**3. Identificar el estadístico de prueba y su distribución**")
print("Para probar una hipótesis sobre la varianza poblacional (σ²), usamos el estadístico Chi-cuadrada (χ²).")
print("Distribución bajo H₀: Distribución Chi-cuadrada con n-1 grados de libertad.")
print("Fórmula del estadístico χ²: χ² = ((n - 1) * s²) / σ²")

# Datos proporcionados
n = 12       # Tamaño de la muestra
s2 = 0.3     # Varianza muestral
sigma2_0 = 0.5 # Varianza hipotetizada (según H0)
grados_libertad = n - 1 # Grados de libertad

# 4. Determine los grados de libertad, el valor crítico, la región crítica y la región de rechazo.
# Para una prueba unilateral a la izquierda con α = 0.05 y n-1 grados de libertad,
# el valor crítico de χ² deja el 5% del área en la cola izquierda.
# chi2.ppf(alpha, grados_libertad) nos da el valor χ² tal que P(χ² <= χ²_critico) = alpha
chi2_critico = chi2.ppf(alpha, grados_libertad)

print(f"\n**4. Determine los grados de libertad, el valor crítico, la región crítica y la región de rechazo.**")
print(f"Grados de libertad (df): {grados_libertad}")
print(f"Valor crítico de χ² para α={alpha} (cola izquierda): {chi2_critico:.4f}")
print(f"Región de rechazo: χ²_calculado ≤ {chi2_critico:.4f}")


# 5. Calcule el estadístico χ² y el valor p.
# Calcular el estadístico χ²
chi2_estadistico = ((n - 1) * s2) / sigma2_0
# Calcular el valor p para una prueba unilateral izquierda: P(χ² <= χ²_estadistico)
valor_p = chi2.cdf(chi2_estadistico, grados_libertad)


print(f"\n**5. Calcule el estadístico χ² y el valor p.**")
print(f"Estadístico χ² calculado: {chi2_estadistico:.4f}")
print(f"Valor p: {valor_p:.4f}")


# 6. Graficar la distribución Chi-cuadrada, región de rechazo, valor crítico y estadístico calculado
print(f"\n**6. Graficar la distribución correspondiente, indicando claramente la región de rechazo, el valor crítico y el estadístico calculado.**")
# Rango de valores para graficar la distribución Chi-cuadrada
# Usamos los valores críticos y el estadístico calculado para definir un rango apropiado para la gráfica
min_val = min(chi2_critico, chi2_estadistico, chi2.ppf(0.001, grados_libertad)) * 0.9 # Asegurar que el rango sea suficiente
x = np.linspace(min_val, chi2.ppf(0.999, grados_libertad), 1000)
y = chi2.pdf(x, grados_libertad) # PDF de la distribución Chi-cuadrada

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'Distribución χ² (df={grados_libertad})')

# Rellenar la región de rechazo (cola izquierda)
x_reject = np.linspace(min_val, chi2_critico, 100)
y_reject = chi2.pdf(x_reject, grados_libertad)
plt.fill_between(x_reject, y_reject, color='red', alpha=0.5, label=f'Región de Rechazo (α={alpha})')

# Marcar el valor crítico de χ²
plt.axvline(chi2_critico, color='red', linestyle='--', label=f'Valor Crítico χ² = {chi2_critico:.4f}')

# Marcar el estadístico χ² calculado
plt.axvline(chi2_estadistico, color='blue', linestyle='-', label=f'Estadístico χ² Calculado = {chi2_estadistico:.4f}')

# Añadir etiquetas y título
plt.title(f'Distribución χ² (df={grados_libertad}) con Región de Rechazo y Estadístico χ² (Prueba Unilateral Izquierda)')
plt.xlabel('Valor de χ²')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Tomar una decisión estadística y concluir en términos prácticos
print(f"\n**7. Tomar una decisión estadística y concluir en términos prácticos**")
print("Para tomar la decisión, comparamos el estadístico χ² calculado con el valor crítico de χ², o el valor p con el nivel de significancia α.")

print(f"\nComparando estadístico χ² vs. valor crítico:")
print(f"- Estadístico χ² calculado: {chi2_estadistico:.4f}")
print(f"- Valor crítico χ²: {chi2_critico:.4f}")
if chi2_estadistico <= chi2_critico:
    print(f"Como {chi2_estadistico:.4f} ≤ {chi2_critico:.4f}, el estadístico χ² calculado cae en la región de rechazo. **Rechazamos la hipótesis nula (H₀)**.")
else:
    print(f"Como {chi2_estadistico:.4f} > {chi2_critico:.4f}, el estadístico χ² calculado NO cae en la región de rechazo. **No rechazamos la hipótesis nula (H₀)**.")

print(f"\nComparando valor p vs. α:")
print(f"- Valor p: {valor_p:.4f}")
print(f"- α: {alpha}")
if valor_p <= alpha:
    print(f"Como {valor_p:.4f} ≤ {alpha}, el valor p es menor que α. **Rechazamos la hipótesis nula (H₀)**.")
else:
     print(f"Como {valor_p:.4f} > {alpha}, el valor p es mayor que α. **No rechazamos la hipótesis nula (H₀)**.")

print("\n**Conclusión en términos prácticos:**")
if valor_p <= alpha: # Usar valor p para la conclusión general para mayor claridad
    print(f"Con un nivel de significancia del {alpha*100}%, hay suficiente evidencia estadística para concluir que la varianza poblacional en el grosor de las piezas metálicas es significativamente menor a 0.5 mm².")
else:
    print(f"Con un nivel de significancia del {alpha*100}%, no hay suficiente evidencia estadística para concluir que la varianza poblacional en el grosor de las piezas metálicas es significativamente menor a 0.5 mm².")

print("\n**Significado del valor p:**")
print(f"El valor p de {valor_p:.4f} significa que, si la varianza poblacional real fuera de 0.5 mm² (es decir, si H₀ fuera verdadera), solo habría un {valor_p*100:.2f}% de probabilidad de observar una varianza muestral de {s2} o algo más extremo (menor que 0.5) simplemente por casualidad en una muestra de {n} piezas.")
if valor_p <= alpha:
     print(f"Dado que esta probabilidad es baja (menor que nuestro umbral α de {alpha}), consideramos que el resultado observado es poco probable bajo H₀, lo que nos lleva a rechazarla.")
else:
    print(f"Dado que esta probabilidad es alta (mayor que nuestro umbral α de {alpha}), consideramos que el resultado observado es probable bajo H₀, lo que nos lleva a no rechazarla.")
```
